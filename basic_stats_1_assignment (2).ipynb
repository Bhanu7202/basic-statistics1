# Descriptive Analytics for Numerical Columns

```python
#Load the dataset into a programming environment
import pandas as pd
df=pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
#Identify numerical columns in the dataset.
numerical_columns = df.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
#Calculate the mean
import statistics as st
for column in df.select_dtypes(include='number').columns:
    mean = st.mean(df[column])
    print(f'mean of {column} is:{mean}')
```

```python
#Calculate the median
for column in df.select_dtypes(include='number').columns:
    median = st.median(df[column])
    print(f'median of {column} is:{median}')
```

```python
#Calculate the mode
for column in df.select_dtypes(include='number').columns:
    mode = st.mode(df[column])
    print(f'mode of {column} is:{mode}')
```

```python
#Calculate the standard deviation
for column in df.select_dtypes(include='number').columns:
    sd = st.stdev(df[column])
    print(f'standard deviation of {column} is:{sd}')
```

# Data Visualization

# histograms

```python
import matplotlib.pyplot as plt
for column in df.select_dtypes(include='number').columns:
    plt.hist(df[column],edgecolor='black',color='grey')
    plt.title(f'Boxplot for {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('frequency')
    plt.show()
```

```python
# finding skewness
from scipy.stats import skew
for column in df.select_dtypes(include='number').columns:
    k= df[column].skew()
    print(f'skew of {column} is:{k}')
```

```python
skewness= df[numerical_columns].skew()
print('skewness of numerical columns is:\n' , skewness)
```

```python
''' A value between -1 & -0.5 (or) between 0.5 & 1 Indicades a moderately skewedÂ distribution

    A value less than - 1.5 (or) grater than 1.5 indicates It is extremely skewedÂ distribution'''
```

```python
from scipy.stats import kurtosis
kurtosis=df[numerical_columns].kurtosis()
print('kurtosis of numerical columnsis:\n',kurtosis)
```

```python
''' A kurtosis grater than 2 suggests too peak distribution, while less than -2 indicates too flat one'''
```

# Bar Chart Analysis for Categorical Column

```python

# Function to plot bar charts for categorical columns
def plot_categorical_distributions(df, categorical_columns):
    for column in categorical_columns:
        df[column].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
        plt.title(f'Distribution of {column}', fontsize=16)
        plt.xlabel(column, fontsize=14)
        plt.ylabel('Count', fontsize=14)
        plt.show()
# Extract categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Plot distributions
plot_categorical_distributions(df, categorical_columns)

```

# Standardization of Numerical Variables

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()
df1=df.copy()
numerical_columns1 = df1.select_dtypes(include=['number']).columns

# Standardize numerical columns
df1[numerical_columns1] = scaler.fit_transform(df1[numerical_columns1])

# Display the standardized dataset
print(df1)
print(df)
```
